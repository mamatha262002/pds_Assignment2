## B. Remove the units from some of the attributes and only keep the numerical values

```python
df1= pd.DataFrame(data)
```

```python
df1['Mileage'] = df1['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df1['Engine'] = df1['Engine'].str.replace(' CC', '').astype(float)

df1['Power'] = df1['Power'].str.extract('(\d+\.\d+)').astype(float)

df1['New_Price'] = df1['New_Price'].str.extract('(\d+\.\d+)').astype(float)

df1.head()
```

# a. Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them.

```python
df1['Mileage'].fillna(df1['Mileage'].median(),inplace=True)
df1['Engine'].fillna(df1['Engine'].mean(),inplace=True)
df1['Power'].fillna(df1['Power'].median(),inplace=True)

df1['Seats'].fillna(df1['Seats'].mode()[0],inplace=True)

df1.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'preprocessed_data.csv'
df1.to_csv(output_csv_file, index=False)
```

# C. Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
df['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
df['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
df['Owner_Type'].replace({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4}, inplace=True)
```
